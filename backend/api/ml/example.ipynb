 ## перевод массива InMemoryUploadedFile в массив PIL

```python
from PIL import Image
from io import BytesIO

def imuf_array_to_pil_array(arr):    
    res = []
    for file in arr:
        file_data = file.read()
        image_data = BytesIO(file_data)
        res.append(Image.open(image_data))
    return res
```
